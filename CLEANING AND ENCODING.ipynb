```python
df=data.copy()
```

```python
cols_label=['Education','EmploymentType','HasMortgage','HasDependents','HasCoSigner']
cols_onehotencoding=['MaritalStatus','LoanPurpose']
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
for col in cols_label:
    df[col]=label_encoder.fit_transform(df[col])
```

```python
df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
hot_encoder=OneHotEncoder()
for col in cols_onehotencoding:
    df[col]=hot_encoder.fit_transform(df[col].values.reshape(-1, 1)).toarray()
```

```python
df.dtypes
```

```python
df[cols_onehotencoding]=df[cols_onehotencoding].astype('int32')
```

```python
df_cleaned=df.copy()
```

```python
df_cleaned.to_csv('cleaned_loan_data.csv',index=False)
```
